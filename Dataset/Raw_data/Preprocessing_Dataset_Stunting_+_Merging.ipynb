```python
data_stunt_2.columns = ['Gender', 'Age', 'Birth Weight', 'Birth Length', 'Body Weight', 'Body Length', 'Breastfeeding', 'Stunting']
data_stunt_2
```

```python
labelencoder = LabelEncoder()
data_stunt['Gender'] = labelencoder.fit_transform(data_stunt['Gender'])
data_stunt['Breastfeeding'] = labelencoder.fit_transform(data_stunt['Breastfeeding'])
data_stunt['Stunting'] = labelencoder.fit_transform(data_stunt['Stunting'])

data_stunt
```

```python
data_stunt_2['Gender'] = labelencoder.fit_transform(data_stunt_2['Gender'])
data_stunt_2['Breastfeeding'] = labelencoder.fit_transform(data_stunt_2['Breastfeeding'])
data_stunt_2['Stunting'] = labelencoder.fit_transform(data_stunt_2['Stunting'])

data_stunt_2
```

```python
merged_df = pd.concat([data_stunt, data_stunt_2], ignore_index=True)
merged_df
```

```python
columns_to_replace = ["Body Weight", "Body Length"]
for column in columns_to_replace:
    merged_df[column] = merged_df[column].astype(str)
    merged_df[column] = merged_df[column].apply(lambda x: x.replace(",", ".") if "," in x else x)
    merged_df[column] = pd.to_numeric(merged_df[column], errors='coerce')
merged_df
```

```python
merged_df.info()
```

```python
merged_df = merged_df.drop(columns=['Birth Weight', 'Birth Length'])
merged_df
```

```python
print("Duplicated: ", merged_df.duplicated().sum())
```

```python
merged_df.drop_duplicates(inplace=True)
```
